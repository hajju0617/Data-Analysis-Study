```r
install.packages("tidyverse")
library(tidyverse)
```

```r
rownames_to_column(mtcars)
```

```r
# tibble 데이터프레임으로 변환.
mtcars1 <- as_tibble(rownames_to_column(mtcars))
mtcars1
```

```r
# ggplot() 함수와 geom 계열의 함수가 '+'로 연결돼야만 기본적인 그래프가 출력됨.
ggplot(mtcars1, aes(hp, mpg, color=factor(cyl)))
```

```r
ggplot(mtcars1, aes(hp, mpg, color=factor(cyl))) + geom_point()
```

```r
ggplot(mtcars1, aes(hp, mpg)) + geom_point(aes(color=factor(cyl)))
```

```r
# group=123 : 범주 구분하지않고 통합해서 시각화 실행.
ggplot(mtcars1, aes(hp, mpg, color=factor(cyl))) + geom_point() + geom_smooth(aes(group=123), se=FALSE)
```

```r
# 부가 정보 추가.
ggplot(mtcars1, aes(hp, mpg, color=factor(cyl)))+ geom_point() +
labs(title="Relationship between Fuel efficiency and Horse power",
      x="Horse Power",
      y="Mile per Gallon",
      caption="Data Source: R", color="Cylinder")
```

```r
ggplot(mtcars1, aes(hp, mpg, color=factor(cyl)))+ geom_point() +
labs(title="Relationship between Fuel efficiency and Horse power",
      x="Horse Power",
      y="Mile per Gallon",
      caption="Data Source: R", color="Cylinder")+
geom_text(aes(label=rowname), show.legend = FALSE)

ggplot(mtcars1, aes(hp, mpg, color=factor(cyl)))+ geom_point() +
labs(title="Relationship between Fuel efficiency and Horse power",
      x="Horse Power",
      y="Mile per Gallon",
      caption="Data Source: R", color="Cylinder")+
geom_text(aes(label=rowname))
```

```r
ggplot(mtcars1, aes(hp, mpg, color=factor(cyl)))+ geom_point() +
labs(title="Relationship between Fuel efficiency and Horse power",
      x="Horse Power",
      y="Mile per Gallon",
      caption="Data Source: R", color="Cylinder")+
geom_label(aes(label=rowname), nudge_x=2.5, nudge_y=1.5, label_size=0.25, show.legend = FALSE)
```

```r
hmpg <- mtcars1 %>% group_by(cyl) %>% filter(row_number(desc(mpg))==1)
hmpg
```

```r
ggplot(mtcars1, aes(hp, mpg, color=factor(cyl)))+ geom_point() +
labs(title="Relationship between Fuel efficiency and Horse power",
      x="Horse Power",
      y="Mile per Gallon",
      caption="Data Source: R", color="Cylinder")+
geom_label(aes(label=rowname), data=hmpg, nudge_x=2.5, nudge_y=1.5, label_size=0.25, show.legend = FALSE)
```

```r
# 각 축의 척도 값, 범례 삭제.
ggplot(mtcars1, aes(hp, mpg, color=factor(cyl)))+ geom_point() +
labs(title="Relationship between Fuel efficiency and Horse power",
      x="Horse Power",
      y="Mile per Gallon",
      caption="Data Source: R", color="Cylinder")+
scale_x_continuous(labels=NULL) + scale_y_continuous(labels=NULL) + scale_color_discrete(labels=NULL)
```

```r
# y축의 연속형 변수의 범위를 10~40으로 지정. (10씩 증가하도록 구간 나눈것.)
ggplot(mtcars1, aes(hp, mpg, color=factor(cyl)))+ geom_point() +
labs(title="Relationship between Fuel efficiency and Horse power",
      x="Horse Power",
      y="Mile per Gallon",
      caption="Data Source: R", color="Cylinder")+
scale_y_continuous(breaks=seq(10, 40, by=10))
```

```r
# x축: 100~250, y축: 10~30까지만해서 출력.
ggplot(mtcars1, aes(hp, mpg, color=factor(cyl)))+ geom_point() +
geom_smooth(aes(group=123), se=FALSE) +
labs(title="Relationship between Fuel efficiency and Horse power",
      x="Horse Power",
      y="Mile per Gallon",
      caption="Data Source: R", color="Cylinder")+
coord_cartesian(xlim=c(100, 250), ylim=c(10, 30))
```

```r
# 범례의 위치 및 크기.
ggplot(mtcars1, aes(hp, mpg, color=factor(cyl)))+ geom_point() +
labs(title="Relationship between Fuel efficiency and Horse power",
      x="Horse Power",
      y="Mile per Gallon",
      caption="Data Source: R", color="Cylinder")+
theme(legend.position = "top")

ggplot(mtcars1, aes(hp, mpg, color=factor(cyl)))+ geom_point() +
labs(title="Relationship between Fuel efficiency and Horse power",
      x="Horse Power",
      y="Mile per Gallon",
      caption="Data Source: R", color="Cylinder")+
theme(legend.position = "bottom")

# 범례를 3행으로 표시.
ggplot(mtcars1, aes(hp, mpg, color=factor(cyl)))+ geom_point() +
labs(title="Relationship between Fuel efficiency and Horse power",
      x="Horse Power",
      y="Mile per Gallon",
      caption="Data Source: R", color="Cylinder")+
theme(legend.position = "bottom") +
guides(color=guide_legend(nrow=3, override.aes=list(size=5)))
```

```r
# 배경 디자인. (theme_classic() : 그리드 생략한 배경)
ggplot(mtcars1, aes(hp, mpg, color=factor(cyl)))+ geom_point() +
labs(title="Relationship between Fuel efficiency and Horse power",
      x="Horse Power",
      y="Mile per Gallon",
      caption="Data Source: R", color="Cylinder")+
theme(legend.position = "bottom") +
guides(color=guide_legend(nrow=3, override.aes=list(size=5))) +
theme_classic() + theme(legend.position = "right")
```
